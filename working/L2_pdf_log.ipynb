```python
import numpy as np
from matplotlib import pyplot as plt
```

N random samples from a uniform distribution

```python
x = np.random.uniform(0.1,10,10000)
```

```python
plt.figure(dpi = 100)
plt.hist(x,density=True,histtype='step',lw=1);
plt.xlabel("x")
plt.show()
```

```python
y = np.log10(x)

plt.figure(dpi=100)
plt.hist(y,density=True,histtype='step',lw=1)
plt.xlabel("y")
```

Using the equation to transform probability distributions, the theoretical pdf of y is: \
y=log10(x) -> x=10^y \
dx/dy=10^y*log(10) \
p(x)=1/(10-0.1) -> p(y)=|dx/dy|*p(x)=10^y*log(10)*1/(10-0.1)

```python
ylin = np.linspace(-1,1,1000)
pdfy = 10**ylin * np.log(10)* 1/(10-0.1)
plt.xlabel('y')
plt.ylabel('p(y)')
plt.plot(ylin, pdfy)
```

```python
print('log mean of x: ', np.log10(np.mean(x))) 
print('mean of y: ', np.mean(y))
print('log median of x: ', np.log10(np.median(x)))
print('median of y: ', np.median(y))
```

The mean changes with the scale of the values because it's influenced by the actual values of the dataset, while the median is invariant under monotonic transformations because it depends solely on the order of the values.
